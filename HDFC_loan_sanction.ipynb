##Data pre-processing

As we that between many features there is data which is very different from each other having larger difference between the means and variance, 
therefore we will use standard scalar which will make every features mean=0 and variance=1, due to which model will not give more importance to a particular feature just because of thier large value

for regression models all the features data should be on same or very close to mean and variance.So that model doesn't give any unnecessary advantage
to other features

StandarScaler works on the principle-:
z=(z-u)/std

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
data=scaler.fit_transform(data)
```

```python
data
```

Model Training Begins

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    data, target, test_size=0.33, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model=LogisticRegression(penalty='l2', C=0.5, tol=0.0001, fit_intercept=True, intercept_scaling=1,
                   class_weight=None, random_state=None, solver='lbfgs', max_iter=300, verbose=0, warm_start=False, n_jobs=None)
```

```python
model.fit(X_train,y_train)
```

```python
y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_predict))
```

Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
```

```python
rf_model = RandomForestClassifier(
    n_estimators=1000,       # 100 trees standard aur stable hain
    max_depth=10,# Kyunki sirf 4 features hain, depth 5 kaafi hai (overfitting se bachega)
    criterion='entropy',
    max_features='sqrt',    # 4 ka square root 2 hota hai, toh har split pe 2 features check honge
    min_samples_split=5,    # Ek node split tabhi hoga jab kam se kam 5 samples hon
    random_state=42,        # Taaki result har baar same aaye
    n_jobs=-1               # Saare CPU cores use karne ke liye (fast processing)
)
```

```python
rf_model.fit(X_train, y_train)
```

```python
y_predict=rf_model.predict(X_test)
```

```python
print(classification_report(y_test,y_predict))
```

```python
importances = rf_model.feature_importances_
```

```python
feature_importance_series = pd.Series(importances,index=sdata.columns)

sorted_importance = feature_importance_series.sort_values(ascending=False)
```

```python
plt.figure(figsize=(10, 6))
sorted_importance.plot(kind='barh', color='skyblue')
plt.title("Feature Importance in Random Forest Model")
plt.xlabel("Feature Importance Score (Mean Decrease in Impurity)")
plt.ylabel("Feature Names")
plt.gca().invert_yaxis() 
plt.show()
```
